```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

### Seperating data for analysis
  - 0 : Normal transaction
  - 1 : Fraudulent transaction

```python
legit = transaction_dataset[transaction_dataset.Class == 0]
fraud = transaction_dataset[transaction_dataset.Class == 1]
```

### Under-Sampling 

- build a sample dataset having similar distribution of normal and fraudulent transactions.
- number of fraudulent transaction is = 492

```python
legit_sample = legit.sample(n = 492)
```

#### Concatinating two samplings

```python
new_transaction_dataset2 = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_transaction_dataset2.head()
```

```python
new_transaction_dataset2.tail()
```

```python
new_transaction_dataset2['Class'].value_counts()
```

```python
new_transaction_dataset2.groupby('Class').mean()
```

### Splitting the data into features and Targets

```python
X = new_transaction_dataset2.drop(columns = 'Class', axis = 1)
Y = new_transaction_dataset2['Class']
```

```python
X
```

```python
Y
```

### Splitting the data into training and testing data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print("Shape of X_train ", X_train.shape)
print("Shape of X_test ", X_test.shape)
print("Shape of Y_train ", Y_train.shape)
print("Shape of Y_test ", Y_test.shape)
```

### Model Training 

# Logistic Regression

```python
model1 = LogisticRegression( max_iter=10000)
model1.fit(X_train, Y_train)

accuracy_results = {}

train_pred = model1.predict(X_train)
train_accuracy = accuracy_score(Y_train, train_pred)

test_pred = model1.predict(X_test)
test_accuracy = accuracy_score(Y_test, test_pred)

accuracy_results["Logistic Regression Train Accuracy"] = train_accuracy
accuracy_results["Logistic Regression Test Accuracy"] = test_accuracy

print(f"Train Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

print("\nClassification Report (Test Data):\n", classification_report(Y_test, test_pred))
print("Confusion Matrix (Test Data):\n", confusion_matrix(Y_test, test_pred))
```

# Decision Tree

```python
model2 = DecisionTreeClassifier(random_state=42)
model2.fit(X_train, Y_train)

accuracy_results = {}

train_pred = model2.predict(X_train)
train_accuracy = accuracy_score(Y_train, train_pred)

test_pred = model2.predict(X_test)
test_accuracy = accuracy_score(Y_test, test_pred)

accuracy_results["Decision Tree Train Accuracy"] = train_accuracy
accuracy_results["Decision Tree Test Accuracy"] = test_accuracy

print(f"Train Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
print("\nClassification Report (Test Data):\n", classification_report(Y_test, test_pred))
print("Confusion Matrix (Test Data):\n", confusion_matrix(Y_test, test_pred))
```

# Random Forest

```python
model3 = RandomForestClassifier(n_estimators=100, random_state=42)
model3.fit(X_train, Y_train)

accuracy_results = {}

train_pred = model3.predict(X_train)
train_accuracy = accuracy_score(Y_train, train_pred)

test_pred = model3.predict(X_test)
test_accuracy = accuracy_score(Y_test, test_pred)

accuracy_results["RandomForest Train Accuracy"] = train_accuracy
accuracy_results["RandomForest Test Accuracy"] = test_accuracy

print(f"Train Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

print("\nClassification Report (Test Data):\n", classification_report(Y_test, test_pred))
print("Confusion Matrix (Test Data):\n", confusion_matrix(Y_test, test_pred))
```

# SVM

```python
model4 = SVC(kernel='rbf', random_state=42)
model4.fit(X_train, Y_train)

accuracy_results = {}

train_pred = model4.predict(X_train)
train_accuracy = accuracy_score(Y_train, train_pred)

test_pred = model4.predict(X_test)
test_accuracy = accuracy_score(Y_test, test_pred)

accuracy_results["SVM Train Accuracy"] = train_accuracy
accuracy_results["SVM Test Accuracy"] = test_accuracy

print(f"Train Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

print("\nClassification Report (Test Data):\n", classification_report(Y_test, test_pred))
print("Confusion Matrix (Test Data):\n", confusion_matrix(Y_test, test_pred))
```
